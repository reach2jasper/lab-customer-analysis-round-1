```python
#Read the three files into python as dataframes
import pandas as pd
import numpy as nm
data1=pd.read_csv('file1.csv')
data2=pd.read_csv('file2.csv')
data3=pd.read_csv('file3.csv')
```

```python
#Show the DataFrame's shape.
print(data1.shape)
print(data2.shape)
print(data3.shape)
```

```python
#Standardize header names.
cols=[]
for column in data1.columns:
    cols.append(column.lower())
cols
data1.columns=cols
data1
cols=[]
for column in data2.columns:
    cols.append(column.lower())
cols
data2.columns=cols
data2
cols=[]
for column in data3.columns:
    cols.append(column.lower())
cols
data3.columns=cols
data3

display(data1.head(0))
display(data2.head(0))
display(data3.head(0))
```

```python
# Rearrange the columns in the dataframe as needed
data2=data2[['customer','st','gender','education','customer lifetime value','income','monthly premium auto',
             'number of open complaints','policy type','vehicle class','total claim amount']]
data3=data3[['customer','state','gender','education','customer lifetime value','income','monthly premium auto',
             'number of open complaints','policy type','vehicle class','total claim amount']]
display(data1.head(0))
display(data2.head(0))
display(data3.head(0))
```

```python
data3=data3.rename(columns={"state":"st"})
data3
```

```python
# Concatenate the three dataframes
master_data=pd.concat([data1,data2,data3],axis=0)
master_data.shape
```

```python
# Which columns are numerical?
master_data._get_numeric_data()
```

```python
# Which columns are categorical?
master_data.select_dtypes('object')
```

```python
# Understand the meaning of all columns
master_data.columns
```

```python
# Perform the data cleaning operations mentioned so far in class
master_data.isna().sum()
```

```python
# Delete the column education and the number of open complaints from the dataframe.
master_data = master_data.drop(['education','number of open complaints'], axis=1)
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
master_data['customer lifetime value'] =  pd.to_numeric(master_data['customer lifetime value'], errors='coerce')
master_data._get_numeric_data()
master_data['customer lifetime value']=master_data['customer lifetime value']/100
master_data['customer lifetime value']
```

```python
# Check for duplicate rows in the data and remove if any.
master_data=master_data.drop_duplicates()
master_data.shape
```

```python
# Filter out the data for customers who have an income of 0 or less.
master_data[master_data['income']<=0]
```
